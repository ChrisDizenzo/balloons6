```python
centerx = 165
centery = 380
width = 60

im = PIL.ImageGrab.grab()
img = np.array(im)
plt.imshow(img[(centery-width):(centery+width),(centerx-width):(centerx+width)])
plt.show()
print(pytesseract.image_to_string(img[(centery-width):(centery+width),(centerx-width):(centerx+width)],
                                  lang='eng',
                                  config='--psm 7 --oem 3 -c tessedit_char_whitelist=0123456789'
                                 ))
```
